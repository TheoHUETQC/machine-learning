## Nettoyage

On doit, avant de l'utiliser, le nettoyer, le '**clean**'.

### Remplacer les Nan

Si on trouve des **Nan** dans notre dataset, il faut verifier le type de la colonne.

Pour les trouver :
```python
df.isna().sum()
```

On ne retire surtout pas les lignes où des Nan sont present pour ne pas perdre de data.

En fonction de quel type de colonne on a, nous allons appliquer une methode differente.
Si on veut juste mettre une valeur specifique on peut quand meme :
```python
df['notre_colonne'].fillna(-1, inplace=True) #remplace tout les nan en -1
```

#### colonne type Objet :
prendre la valeur au dessus (ou en dessous).
sur Python :
- `.bfill()` = valeur d'apres
- `.ffill()` = valeur d'avant

```python
df['notre_colonne'].ffill(inplace=True) # ou bfill
```

#### colonne type int ou float :
prendre la **medianne** ou **moyenne**.
sur python :
```python
df['notre_colonne'].fillna(df['notre_colonne'].median(), inplace = True) # ou .mean() pour la moyenne
```

### Transformer les Objets en int/float

Certaine colonnes colonnent peuvent contenir des objets qui sont en réalité des nombres.
Par exemple une colonne de prix `34000€', il faut retirer le "€" :
```python
# Replace the $ by nothing
df['notre_colonne']= df['notre_colonne'].str.replace('$','')
     
# Change the type of the column "Fare" from object(string) to float
df['notre_colonne']= df['notre_colonne'].astype(float)
```

D'autre colonne peuvent etre des Booleens, qui peuvent etre transformer en 0, 1.
Par exemple une colonne de 'Yes', 'No' :
```python
df['notre_colonne']= df['notre_colonne'].replace("Yes", 1)
df['notre_colonne']= df['notre_colonne'].replace("No", 0)
```

## Standard Scaler

On veut des **moyenne = 0** et des **ecart types = 1** car la machine sera plus performante.
On peut donc changer l'echelles de nos valeurs pour avoir ses valeurs

On applique une fonction linéaire = **rescale** pour finalement avoir :
- standard deviation = 1
- mean = 0
dans le data set de train

On apllique tout de meme le rescale sur le dataset de test.

**Remarque :**
- Nos valeurs seront sans units
- Ce n'est pas pour tout les modèles (ca ne marche pas si on veut comparer a une valeur seuil).

### Formule scaler :

$ z = \frac{x - \mu_{train}}{\sigma_{train}} $

avec :
- $x =$ notre valeur rélle que l'on veut rescale
- $\mu_{train} =$ la moyenne réelle
- $\sigma_{train} =$ l'écart type réelle

### Sur Python :

On utilise `.fit()` qui permet de calculer directement la moyenne...

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

# on retourne la data transformé
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# on retransforme en data frame
X_train_df = pd.DataFrame(X_train_scaled)
X_test_df = pd.DataFrame(X_test_scaled)
````
