## Standard Scaler

On veut des **moyenne = 0** et des **ecart types = 1** car la machine sera plus performante.
On peut donc changer l'echelles de nos valeurs pour avoir ses valeurs

On applique une fonction linéaire = **rescale** pour finalement avoir :
- standard deviation = 1
- mean = 0
dans le data set de train

On apllique tout de meme le rescale sur le dataset de test.

**Remarque :**
- Nos valeurs seront sans units
- Ce n'est pas pour tout les modèles (ca ne marche pas si on veut comparer a une valeur seuil).

### Formule scaler :

$ z = \frac{x - \mu_{train}}{\sigma_{train}} $

avec :
- $x =$ notre valeur rélle que l'on veut rescale
- $\mu_{train} =$ la moyenne réelle
- $\sigma_{train} =$ l'écart type réelle

On utilise `.fit()` en vrai qui permet de calculer ma moyenne... directement

### Sur Python :
```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

# on retourne la data transformé
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```
